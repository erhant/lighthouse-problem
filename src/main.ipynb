Now let us plot how the darkness changes as the lighthouse number grows.


```python
maxL = 30
N_L = range(1,maxL+1)
DA, DA_theorem = zip(*[compute_1(N, print_res=False) for N in N_L]) 
fig = plt.figure()
ax = plt.axes()
ax.plot(N_L[::2], DA_theorem[::2], c='green')
ax.plot(N_L[::2], DA[::2], c='blue')
errs = [abs(DA[i]-DA_theorem[i]) for i in range(0, maxL, 2)]
ax.plot(N_L[::2], errs, c='red')
ax.set_xlabel("Number of Lighthouses")
ax.set_ylabel("Total Dark Area")
plt.show()
```

```python
maxL = 100
N_L = range(1,maxL+1)
DA = [compute_2(N, print_res=False) for N in N_L]
fig = plt.figure()
ax = plt.axes()
ax.scatter(N_L, DA, c='blue')

# lets find when the area increases
N_L_increase = [i+1 for i in range(2,len(DA)-1) if DA[i] < DA[i+1]]
ax.scatter(N_L_increase, [0.0] * len(N_L_increase), c='orange', marker='x')
ax.set_xlabel("Number of Lighthouses")
ax.set_ylabel("Total Dark Area")
print(N_L_increase)
plt.show()
```
